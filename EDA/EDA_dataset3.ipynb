```python
# Gr√°fico Pie Chart para la distribuci√≥n por tipo de contenido
fig = px.pie(
    df,
    names="type",
    title="Distribuci√≥n de Contenido por Tipo (Pel√≠cula vs Serie)",
    hole=0.3
)
fig.update_traces(textinfo="percent+label")
fig.update_layout(title_font_size=18)
fig.show()
```

El gr√°fico muestra que la mayor√≠a del contenido en el dataset son pel√≠culas (69.6%), mientras que el resto son series de TV (30.4%).

```python
# Histograma de Distribuci√≥n de A√±o de Estreno por Tipo de Contenido
fig = px.histogram(
    df,
    x="release_year",
    color="type",
    nbins=7,
    title="Distribuci√≥n de A√±o de Estreno por Tipo de Contenido"
)
fig.update_layout(bargap=0.1, title_font_size=18)
fig.show()
```

El histograma muestra la cantidad de contenidos estrenados por a√±o, separado entre pel√≠culas y series.

Se puede observar c√≥mo la cantidad de estrenos var√≠a en el tiempo y qu√© tipo de contenido predomina en cada periodo.
