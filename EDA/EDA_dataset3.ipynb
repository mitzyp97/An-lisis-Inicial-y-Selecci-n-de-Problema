```python
# Gráfico Pie Chart para la distribución por tipo de contenido
fig = px.pie(
    df,
    names="type",
    title="Distribución de Contenido por Tipo (Película vs Serie)",
    hole=0.3
)
fig.update_traces(textinfo="percent+label")
fig.update_layout(title_font_size=18)
fig.show()
```

El gráfico muestra que la mayoría del contenido en el dataset son películas (69.6%), mientras que el resto son series de TV (30.4%).

```python
# Histograma de Distribución de Año de Estreno por Tipo de Contenido
fig = px.histogram(
    df,
    x="release_year",
    color="type",
    nbins=7,
    title="Distribución de Año de Estreno por Tipo de Contenido"
)
fig.update_layout(bargap=0.1, title_font_size=18)
fig.show()
```

El histograma muestra la cantidad de contenidos estrenados por año, separado entre películas y series.

Se puede observar cómo la cantidad de estrenos varía en el tiempo y qué tipo de contenido predomina en cada periodo.

No se pudo realizar el mapa de calor de correlación porque solo existe una variable numérica directa en el dataset (release_year).
